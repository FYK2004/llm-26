> 1) Based on the training dataset (collect all sentences in training dataset), build unigram, bigram, and trigram language models using Additive smoothing technique. It is encouraged to implement models by yourself.


```python
# Your code goes to here
import numpy as np
import math
import re
from collections import Counter
from typing import List, Tuple, Dict, Iterable

# 定义特殊标记
BOS = "<s>"  # 句子开始
EOS = "</s>"  # 句子结束
UNK = "<UNK>"  # 未知词

# 检查预处理数据是否存在
if 'preprocessed_train_data' not in locals() and 'preprocessed_train_data' not in globals():
    print("错误：未找到预处理数据，请先运行Task1的第四部分")
else:
    print(f"找到预处理数据，共 {len(preprocessed_train_data)} 个文档")
    
    # 收集所有句子（将整个文档视为一个长句子）
    all_sentences = [doc['preprocessed_text'] for doc in preprocessed_train_data]
    print(f"收集到 {len(all_sentences)} 个句子（文档）")
    
    # 基于lecture-02的方式构建词汇表
    def build_vocab(sentences: List[List[str]], min_freq: int = 10) -> Dict[str, int]:
        """构建词汇表 """
        # 统计词频
        counts = Counter()
        for sentence in sentences:
            counts.update(sentence)
        
        print(f"总词汇数（去重前）: {len(counts)}")
        
        # 保留频率达到阈值的单词
        words = [w for w, c in counts.items() if c >= min_freq]
        
        # 构建词汇表字典
        vocab = {UNK: 0, BOS: 1, EOS: 2}
        for w in words:
            if w not in vocab:
                vocab[w] = len(vocab)
        
        print(f"词汇表大小（频率≥{min_freq}）: {len(vocab)}")
        print(f"低频词数量: {len(counts) - len(words)}")
        
        return vocab
    
    def map_oov(sents: List[List[str]], vocab: Dict[str, int]) -> List[List[str]]:
        """将词汇表外的词映射为UNK"""
        vset = set(vocab.keys())
        return [[w if w in vset else UNK for w in s] for s in sents]
    
    def add_boundaries(s: List[str], n: int) -> List[str]:
        """添加边界标记"""
        if n == 3:
            return [BOS, BOS] + s + [EOS]
        else:
            return [BOS] + s + [EOS]
    
    def count_ngrams(sents: List[List[str]], n: int) -> Tuple[Counter, Counter]:
        context_counts = Counter()
        ngram_counts = Counter()
        
        for s in sents:
            toks = add_boundaries(s, n)
            if n == 1:
                # unigram: 统计所有单词
                for w in toks:
                    context_counts[()] += 1  # 空上下文
                    ngram_counts[(w,)] += 1
            else:
                # bigram/trigram: 统计ngram
                for i in range(n - 1, len(toks)):
                    ctx = tuple(toks[i - (n - 1): i])  # 上下文
                    ng = tuple(toks[i - (n - 1): i] + [toks[i]])  # ngram
                    context_counts[ctx] += 1
                    ngram_counts[ng] += 1
        
        return context_counts, ngram_counts
    
    # 构建ngram语言模型类
    class NGramLM:
        def __init__(self, n: int, vocab: Dict[str, int], context_counts: Counter, ngram_counts: Counter, alpha: float = 1.0):
            self.n = n
            self.vocab = vocab
            self.V = len(vocab)
            self.context_counts = context_counts
            self.ngram_counts = ngram_counts
            self.alpha = alpha  # additive平滑参数
        
        def probability(self, *words):
            """计算ngram概率"""
            if self.n == 1:
                # unigram概率
                w = words[0]
                c_w = self.ngram_counts.get((w,), 0)
                N = self.context_counts.get((), 0)
                return (c_w + self.alpha) / (N + self.alpha * self.V)
            elif self.n == 2:
                # bigram概率
                w1, w2 = words
                c_w1w2 = self.ngram_counts.get((w1, w2), 0)
                c_w1 = self.context_counts.get((w1,), 0)
                return (c_w1w2 + self.alpha) / (c_w1 + self.alpha * self.V)
            elif self.n == 3:
                # trigram概率
                w1, w2, w3 = words
                c_w1w2w3 = self.ngram_counts.get((w1, w2, w3), 0)
                c_w1w2 = self.context_counts.get((w1, w2), 0)
                return (c_w1w2w3 + self.alpha) / (c_w1w2 + self.alpha * self.V)
        
        def log_probability(self, *words):
            """计算对数概率"""
            prob = self.probability(*words)
            return math.log(max(prob, 1e-10))  # 避免log(0)
        
        def perplexity(self, test_sentences: List[List[str]]) -> float:
            """计算困惑度"""
            total_logprob = 0.0
            total_tokens = 0
            
            for s in test_sentences:
                toks = add_boundaries(s, self.n)
                
                if self.n == 1:
                    # unigram: 预测每个token
                    T = len(toks)
                    for w in toks:
                        total_logprob += self.log_probability(w)
                else:
                    # bigram/trigram: 预测从第n-1个位置开始的token
                    T = len(toks) - (self.n - 1)
                    for i in range(self.n - 1, len(toks)):
                        if self.n == 2:
                            w1, w2 = toks[i-1], toks[i]
                            total_logprob += self.log_probability(w1, w2)
                        elif self.n == 3:
                            w1, w2, w3 = toks[i-2], toks[i-1], toks[i]
                            total_logprob += self.log_probability(w1, w2, w3)
                
                total_tokens += T
            
            if total_tokens > 0:
                return math.exp(-total_logprob / total_tokens)
            return float('inf')
        
        def generate_sentence(self, max_length: int = 15) -> str:
            """生成句子"""
            if self.n == 1:
                # unigram生成
                words = list(self.vocab.keys())
                probs = [self.probability(w) for w in words]
                probs = np.array(probs) / sum(probs)
                
                sentence = []
                for _ in range(max_length):
                    word = np.random.choice(words, p=probs)
                    if word == EOS:
                        break
                    if word != BOS:
                        sentence.append(word)
                return ' '.join(sentence)
            
            elif self.n == 2:
                # bigram生成
                sentence = [BOS]
                current_word = BOS
                
                for _ in range(max_length):
                    # 获取可能的下一单词
                    candidates = []
                    for word in self.vocab:
                        if word not in [BOS, EOS]:
                            prob = self.probability(current_word, word)
                            candidates.append((word, prob))
                    
                    if not candidates:
                        break
                    
                    # 按概率选择
                    words, probs = zip(*candidates)
                    probs = np.array(probs)
                    probs = probs / probs.sum()
                    
                    next_word = np.random.choice(words, p=probs)
                    
                    if next_word == EOS:
                        break
                    
                    sentence.append(next_word)
                    current_word = next_word
                
                return ' '.join(sentence[1:])  # 移除BOS
            
            elif self.n == 3:
                # trigram生成
                sentence = [BOS, BOS]
                
                for _ in range(max_length):
                    w1, w2 = sentence[-2], sentence[-1]
                    
                    candidates = []
                    for word in self.vocab:
                        if word not in [BOS, EOS]:
                            prob = self.probability(w1, w2, word)
                            candidates.append((word, prob))
                    
                    if not candidates:
                        break
                    
                    words, probs = zip(*candidates)
                    probs = np.array(probs)
                    probs = probs / probs.sum()
                    
                    next_word = np.random.choice(words, p=probs)
                    
                    if next_word == EOS:
                        break
                    
                    sentence.append(next_word)
                
                return ' '.join(sentence[2:])  # 移除前两个BOS
    
    def train_ngram_lm(train_sents: List[List[str]], n: int, vocab: Dict[str, int], alpha: float = 1.0) -> NGramLM:
        """训练ngram语言模型"""
        train_sents_mapped = map_oov(train_sents, vocab)
        context_counts, ngram_counts = count_ngrams(train_sents_mapped, n)
        return NGramLM(n=n, vocab=vocab, context_counts=context_counts, ngram_counts=ngram_counts, alpha=alpha)
    
    print("\n=== 构建词汇表 ===")
    vocab = build_vocab(all_sentences, min_freq=10)
    
    # 映射OOV词
    processed_sentences = map_oov(all_sentences, vocab)
    
    print(f"平均句子长度: {np.mean([len(s) for s in processed_sentences]):.2f} 个单词")
    
    # 显示词汇表统计
    print("\n=== 词汇表统计 ===")
    print(f"词汇表大小: {len(vocab)}")
    print(f"特殊标记: {list(vocab.keys())[:3]}")
    
    # 显示处理后的句子示例
    print("\n=== 处理后的句子示例 ===")
    for i in range(min(2, len(processed_sentences))):
        sample_sentence = ' '.join(processed_sentences[i][:15]) + "..."
        print(f"句子 {i+1}: {sample_sentence}")
    
    print("\n=== 开始训练语言模型 ===")
    
    # 训练Unigram模型
    print("1. 训练Unigram模型...")
    unigram_model = train_ngram_lm(processed_sentences, n=1, vocab=vocab, alpha=0.1)
    print("✓ Unigram模型训练完成")
    
    # 训练Bigram模型
    print("2. 训练Bigram模型...")
    bigram_model = train_ngram_lm(processed_sentences, n=2, vocab=vocab, alpha=0.1)
    print("✓ Bigram模型训练完成")
    
    # 训练Trigram模型
    print("3. 训练Trigram模型...")
    trigram_model = train_ngram_lm(processed_sentences, n=3, vocab=vocab, alpha=0.1)
    print("✓ Trigram模型训练完成")
    
    print("\n=== 语言模型训练完成 ===")
    print(f"Unigram模型词汇表大小: {unigram_model.V}")
    print(f"Bigram模型词汇表大小: {bigram_model.V}")
    print(f"Trigram模型词汇表大小: {trigram_model.V}")
    
    # 测试模型
    print("\n=== 模型测试示例 ===")
    test_words = ['the', 'great', 'gatsby', 'is', 'a', 'novel']
    
    print("Unigram概率示例:")
    for word in test_words[:3]:
        prob = unigram_model.probability(word)
        print(f"P({word}) = {prob:.6f}")
    
    print("\nBigram概率示例:")
    for i in range(min(2, len(test_words)-1)):
        w1, w2 = test_words[i], test_words[i+1]
        prob = bigram_model.probability(w1, w2)
        print(f"P({w2}|{w1}) = {prob:.6f}")
    
    print("\nTrigram概率示例:")
    for i in range(min(2, len(test_words)-2)):
        w1, w2, w3 = test_words[i], test_words[i+1], test_words[i+2]
        prob = trigram_model.probability(w1, w2, w3)
        print(f"P({w3}|{w1},{w2}) = {prob:.6f}")
    
    # 生成句子示例
    print("\n=== 生成句子示例 ===")
    print("Unigram生成:")
    for i in range(2):
        sentence = unigram_model.generate_sentence()
        print(f"句子 {i+1}: {sentence}")
    
    print("\nBigram生成:")
    for i in range(2):
        sentence = bigram_model.generate_sentence()
        print(f"句子 {i+1}: {sentence}")
    
    print("\nTrigram生成:")
    for i in range(2):
        sentence = trigram_model.generate_sentence()
        print(f"句子 {i+1}: {sentence}")
```

```python
# Your code goes to here
import numpy as np
from collections import defaultdict, Counter
from typing import List, Dict, Tuple
import math


# 检查预处理数据是否存在
if 'preprocessed_train_data' not in locals() and 'preprocessed_train_data' not in globals():
    print("错误：未找到预处理训练数据，请先运行Task1的第三部分")
else:
    print(f"找到预处理训练数据，共 {len(preprocessed_train_data)} 个文档")
    
    # 检查测试数据是否存在
    if 'preprocessed_test_data' not in locals() and 'preprocessed_test_data' not in globals():
        print("错误：未找到预处理测试数据，请先运行Task1的第四部分")
    else:
        print(f"找到预处理测试数据，共 {len(preprocessed_test_data)} 个文档")
        
        # 朴素贝叶斯分类器实现
        class NaiveBayesClassifier:
            def __init__(self, alpha: float = 1.0):
                self.alpha = alpha
                self.class_priors = {}  # 类别先验概率 P(c)
                self.class_word_counts = {}  # 每个类别中每个单词的计数
                self.class_total_words = {}  # 每个类别的总单词数
                self.vocab = set()  # 词汇表
                self.classes = []  # 所有类别
            
            def fit(self, train_data: List[Dict]):
                print("开始训练朴素贝叶斯分类器...")
                
                # 收集所有类别
                self.classes = list(set([doc['label'] for doc in train_data]))
                print(f"发现类别: {self.classes}")
                
                # 初始化计数器
                class_doc_counts = Counter()
                self.class_word_counts = {c: defaultdict(int) for c in self.classes}
                self.class_total_words = {c: 0 for c in self.classes}
                
                # 统计文档和单词计数
                for doc in train_data:
                    label = doc['label']
                    text = doc['preprocessed_text']
                    
                    class_doc_counts[label] += 1
                    
                    for word in text:
                        self.class_word_counts[label][word] += 1
                        self.class_total_words[label] += 1
                        self.vocab.add(word)
                
                # 计算类别先验概率 P(c)
                total_docs = len(train_data)
                self.class_priors = {c: count / total_docs for c, count in class_doc_counts.items()}
                
                # 计算词汇表大小
                self.vocab_size = len(self.vocab)
                print(f"词汇表大小: {self.vocab_size}")
                print(f"类别分布: {dict(class_doc_counts)}")
                
                print("✓ 朴素贝叶斯分类器训练完成")
            
            def predict_proba(self, text: List[str]) -> Dict[str, float]:

                log_probs = {}
                
                for c in self.classes:
                    # 计算对数先验概率 log P(c)
                    log_prior = math.log(self.class_priors[c])
                    
                    # 计算对数似然概率 sum(log P(w|c))
                    log_likelihood = 0.0
                    for word in text:
                        # 计算单词在类别c中的条件概率 P(w|c)
                        word_count = self.class_word_counts[c].get(word, 0)
                        total_words = self.class_total_words[c]
                        
                        # 使用Laplace平滑
                        p_word_given_class = (word_count + self.alpha) / (total_words + self.alpha * self.vocab_size)
                        log_likelihood += math.log(p_word_given_class)
                    
                    # 对数后验概率 = log P(c) + sum(log P(w|c))
                    log_probs[c] = log_prior + log_likelihood
                
                return log_probs
            
            def predict(self, text: List[str]) -> str:
                log_probs = self.predict_proba(text)
                # 选择具有最高对数概率的类别
                predicted_class = max(log_probs.items(), key=lambda x: x[1])[0]
                return predicted_class
            
            def evaluate(self, test_data: List[Dict]) -> Dict[str, float]:
                print("开始评估朴素贝叶斯分类器...")
                
                y_true = []
                y_pred = []
                
                for doc in test_data:
                    true_label = doc['label']
                    text = doc['preprocessed_text']
                    
                    predicted_label = self.predict(text)
                    
                    y_true.append(true_label)
                    y_pred.append(predicted_label)
                
                # 计算准确率
                accuracy = np.mean(np.array(y_true) == np.array(y_pred))
                
                # 计算每个类别的精确率、召回率、F1分数
                from sklearn.metrics import classification_report, confusion_matrix
                
                print("\n=== 分类报告 ===")
                print(classification_report(y_true, y_pred, target_names=self.classes))
                
                # 计算混淆矩阵
                cm = confusion_matrix(y_true, y_pred)
                print("=== 混淆矩阵 ===")
                print(cm)
                
                # 计算Micro-F1和Macro-F1
                from sklearn.metrics import f1_score
                micro_f1 = f1_score(y_true, y_pred, average='micro')
                macro_f1 = f1_score(y_true, y_pred, average='macro')
                
                results = {
                    'accuracy': accuracy,
                    'micro_f1': micro_f1,
                    'macro_f1': macro_f1,
                    'confusion_matrix': cm
                }
                
                return results
        
        # 训练朴素贝叶斯分类器
        print("\n=== 训练朴素贝叶斯分类器 ===")
        
        # 使用不同的平滑参数进行实验
        alphas = [0.1]# 可以添加其他平滑参数进行实验
        best_alpha = None
        best_macro_f1 = 0
        best_model = None
        
        for alpha in alphas:
            print(f"\n使用平滑参数 alpha = {alpha}")
            
            # 创建并训练模型
            nb_classifier = NaiveBayesClassifier(alpha=alpha)
            nb_classifier.fit(preprocessed_train_data)
            
            # 在测试集上评估
            results = nb_classifier.evaluate(preprocessed_test_data)
            
            print(f"准确率: {results['accuracy']:.4f}")
            print(f"Micro-F1: {results['micro_f1']:.4f}")
            print(f"Macro-F1: {results['macro_f1']:.4f}")
            
            # 选择最佳模型
            if results['macro_f1'] > best_macro_f1:
                best_macro_f1 = results['macro_f1']
                best_alpha = alpha
                best_model = nb_classifier
        
        print(f"\n=== 最佳模型选择 ===")
        print(f"最佳平滑参数: alpha = {best_alpha}")
        print(f"最佳Macro-F1: {best_macro_f1:.4f}")
        
        # 使用最佳模型进行详细分析
        print("\n=== 最佳模型详细分析 ===")
        final_results = best_model.evaluate(preprocessed_test_data)
        
        # 显示一些预测示例
        print("\n=== 预测示例 ===")
        test_samples = preprocessed_test_data[:5]
        for i, sample in enumerate(test_samples):
            true_label = sample['label']
            text = sample['preprocessed_text']
            predicted_label = best_model.predict(text)
     
            print(f"示例 {i+1}:")
            print(f"  真实标签: {true_label}")
            print(f"  预测标签: {predicted_label}")
            print(f"  文本预览: {' '.join(text[:10])}...")
            print()
        
```

```python
# Your code goes to here
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List, Dict, Tuple
import warnings
warnings.filterwarnings('ignore')

print("=== Task3 第二部分：构建逻辑回归分类器 ===")

# 检查预处理数据是否存在
if 'preprocessed_train_data' not in locals() and 'preprocessed_train_data' not in globals():
    print("错误：未找到预处理训练数据，请先运行Task1的第三部分")
else:
    print(f"找到预处理训练数据，共 {len(preprocessed_train_data)} 个文档")
    
    # 检查测试数据是否存在
    if 'preprocessed_test_data' not in locals() and 'preprocessed_test_data' not in globals():
        print("错误：未找到预处理测试数据，请先运行Task1的第四部分")
    else:
        print(f"找到预处理测试数据，共 {len(preprocessed_test_data)} 个文档")
        
        # 准备数据
        print("\n=== 数据准备 ===")
        
        # 提取文本和标签
        train_texts = [' '.join(doc['preprocessed_text']) for doc in preprocessed_train_data]
        train_labels = [doc['label'] for doc in preprocessed_train_data]
        
        test_texts = [' '.join(doc['preprocessed_text']) for doc in preprocessed_test_data]
        test_labels = [doc['label'] for doc in preprocessed_test_data]
        
        print(f"训练集大小: {len(train_texts)}")
        print(f"测试集大小: {len(test_texts)}")
        
        # 编码标签
        label_encoder = LabelEncoder()
        y_train = label_encoder.fit_transform(train_labels)
        y_test = label_encoder.transform(test_labels)
        
        print(f"类别编码: {dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))}")
        
        # 特征工程：尝试不同的特征提取方法
        print("\n=== 特征工程 ===")
        
        def create_features(texts: List[str], method: str = 'tfidf', **kwargs):
            """
            创建文本特征
            Args:
                texts: 文本列表
                method: 特征提取方法 ('tfidf', 'count', 'binary')
            Returns:
                特征矩阵和特征提取器
            """
            if method == 'tfidf':
                vectorizer = TfidfVectorizer(
                    max_features=kwargs.get('max_features', 5000),
                    min_df=kwargs.get('min_df', 2),
                    max_df=kwargs.get('max_df', 0.8),
                    ngram_range=kwargs.get('ngram_range', (1, 1))
                )
            elif method == 'count':
                vectorizer = CountVectorizer(
                    max_features=kwargs.get('max_features', 5000),
                    min_df=kwargs.get('min_df', 2),
                    max_df=kwargs.get('max_df', 0.8),
                    ngram_range=kwargs.get('ngram_range', (1, 1))
                )
            elif method == 'binary':
                vectorizer = CountVectorizer(
                    max_features=kwargs.get('max_features', 5000),
                    min_df=kwargs.get('min_df', 2),
                    max_df=kwargs.get('max_df', 0.8),
                    binary=True,
                    ngram_range=kwargs.get('ngram_range', (1, 1))
                )
            else:
                raise ValueError("不支持的 feature_method")
            
            X = vectorizer.fit_transform(texts)
            return X, vectorizer
        
        # 尝试不同的特征组合
        feature_configs = [
            {'method': 'tfidf', 'max_features': 3000, 'ngram_range': (1, 1), 'name': 'TF-IDF Unigram'},
            {'method': 'tfidf', 'max_features': 5000, 'ngram_range': (1, 2), 'name': 'TF-IDF Bigram'},
            {'method': 'count', 'max_features': 3000, 'ngram_range': (1, 1), 'name': 'Count Unigram'},
            {'method': 'binary', 'max_features': 3000, 'ngram_range': (1, 1), 'name': 'Binary Unigram'}
        ]
        
        # 模型训练和评估
        print("\n=== 逻辑回归模型训练 ===")
        
        best_model = None
        best_vectorizer = None
        best_config = None
        best_score = 0
        best_results = {}
        
        for config in feature_configs:
            print(f"\n--- 使用特征: {config['name']} ---")
            
            # 创建特征
            X_train, vectorizer = create_features(train_texts, **config)
            X_test = vectorizer.transform(test_texts)
            
            print(f"特征维度: {X_train.shape[1]}")
            print(f"特征稀疏度: {X_train.nnz / (X_train.shape[0] * X_train.shape[1]):.4f}")
            
            # 分割训练集用于参数调优
            X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
                X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
            )
            
            # 定义逻辑回归模型
            lr_model = LogisticRegression(
                random_state=42,
                max_iter=1000,
                verbose=0
            )
            
            # 参数网格搜索
            param_grid = {
                'C': [0.1, 1, 10, 100],
                'penalty': ['l2'],  # 对于多分类，l1需要liblinear或saga
                'solver': ['lbfgs', 'liblinear']
            }
            
            print("进行参数调优...")
            grid_search = GridSearchCV(
                lr_model, param_grid, cv=3, scoring='f1_macro', n_jobs=-1, verbose=0
            )
            grid_search.fit(X_train_split, y_train_split)
            
            # 使用最佳参数训练最终模型
            best_params = grid_search.best_params_
            print(f"最佳参数: {best_params}")
            
            final_model = LogisticRegression(**best_params, random_state=42, max_iter=1000)
            final_model.fit(X_train, y_train)
            
            # 在测试集上评估
            y_pred = final_model.predict(X_test)
            
            # 计算评估指标
            accuracy = np.mean(y_pred == y_test)
            micro_f1 = f1_score(y_test, y_pred, average='micro')
            macro_f1 = f1_score(y_test, y_pred, average='macro')
            
            print(f"准确率: {accuracy:.4f}")
            print(f"Micro-F1: {micro_f1:.4f}")
            print(f"Macro-F1: {macro_f1:.4f}")
            
            # 保存最佳模型
            if macro_f1 > best_score:
                best_score = macro_f1
                best_model = final_model
                best_vectorizer = vectorizer
                best_config = config
                best_results = {
                    'accuracy': accuracy,
                    'micro_f1': micro_f1,
                    'macro_f1': macro_f1,
                    'y_pred': y_pred,
                    'config': config
                }
        
        # 使用最佳模型进行详细分析
        print(f"\n=== 最佳模型结果: {best_config['name']} ===")
        print(f"最佳Macro-F1: {best_score:.4f}")
        
        # 详细分类报告
        print("\n=== 详细分类报告 ===")
        print(classification_report(y_test, best_results['y_pred'], 
                                   target_names=label_encoder.classes_))
        
        # 混淆矩阵可视化
        plt.figure(figsize=(10, 8))
        cm = confusion_matrix(y_test, best_results['y_pred'])
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=label_encoder.classes_,
                   yticklabels=label_encoder.classes_)
        plt.title('Confusion Matrix - Logistic Regression')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.tight_layout()
        plt.show()
        
        # 特征重要性分析
        print("\n=== 特征重要性分析 ===")
        
        # 获取特征名称
        feature_names = best_vectorizer.get_feature_names_out()
        
        # 分析每个类别的特征权重
        for class_idx, class_name in enumerate(label_encoder.classes_):
            print(f"\n类别 '{class_name}' 的最重要特征:")
            
            # 获取该类别的系数
            coef = best_model.coef_[class_idx]
            
            # 获取最重要的特征（正权重）
            top_positive_indices = np.argsort(coef)[-10:][::-1]
            top_positive_features = [(feature_names[i], coef[i]) for i in top_positive_indices]
            
            print("  正向特征（支持该类别）:")
            for feature, weight in top_positive_features:
                print(f"    {feature}: {weight:.4f}")
            
            # 获取最重要的负特征（负权重）
            top_negative_indices = np.argsort(coef)[:10]
            top_negative_features = [(feature_names[i], coef[i]) for i in top_negative_indices]
            
            print("  负向特征（反对该类别）:")
            for feature, weight in top_negative_features:
                print(f"    {feature}: {weight:.4f}")
        
        # 模型比较分析
        print("\n=== 模型比较分析 ===")
        
        # 如果有朴素贝叶斯的结果，进行比较
        if 'best_macro_f1' in locals() or 'best_macro_f1' in globals():
            print("与朴素贝叶斯分类器比较:")
            print(f"朴素贝叶斯 Macro-F1: {best_macro_f1:.4f}")
            print(f"逻辑回归 Macro-F1: {best_score:.4f}")
            
            if best_score > best_macro_f1:
                print("逻辑回归性能优于朴素贝叶斯")
            else:
                print("朴素贝叶斯性能优于逻辑回归")
        
        # 错误分析
        print("\n=== 错误分析 ===")
        
        # 找出分类错误的样本
        incorrect_indices = np.where(y_test != best_results['y_pred'])[0]
        print(f"错误分类样本数: {len(incorrect_indices)}")
        
        if len(incorrect_indices) > 0:
            print("\n错误分类示例:")
            for i in range(min(3, len(incorrect_indices))):
                idx = incorrect_indices[i]
                true_label = label_encoder.inverse_transform([y_test[idx]])[0]
                pred_label = label_encoder.inverse_transform([best_results['y_pred'][idx]])[0]
                text_preview = test_texts[idx][:100] + "..." if len(test_texts[idx]) > 100 else test_texts[idx]
                
                print(f"示例 {i+1}:")
                print(f"  真实标签: {true_label}")
                print(f"  预测标签: {pred_label}")
                print(f"  文本预览: {text_preview}")
                print()
        
        # 模型部署和预测函数
        def predict_lr(text: str) -> Tuple[str, Dict[str, float]]:
            """
            使用训练好的逻辑回归模型预测新文本
            Args:
                text: 输入文本
            Returns:
                预测标签和各类别概率
            """
            # 预处理文本（假设已经预处理）
            # 在实际应用中，这里应该包含与训练时相同的预处理步骤
            processed_text = ' '.join(text.split()[:100])  # 简化处理
            
            # 特征提取
            X = best_vectorizer.transform([processed_text])
            
            # 预测
            predicted_class_idx = best_model.predict(X)[0]
            predicted_class = label_encoder.inverse_transform([predicted_class_idx])[0]
            
            # 概率分布
            probabilities = best_model.predict_proba(X)[0]
            prob_dict = {label_encoder.classes_[i]: prob for i, prob in enumerate(probabilities)}
            
            return predicted_class, prob_dict
        
        # 测试预测函数
        print("\n=== 预测函数测试 ===")
        test_samples = test_texts[:3]
        for i, sample in enumerate(test_samples):
            pred_label, probs = predict_lr(sample)
            true_label = label_encoder.inverse_transform([y_test[i]])[0]
            
            print(f"测试样本 {i+1}:")
            print(f"  真实标签: {true_label}")
            print(f"  预测标签: {pred_label}")
            print(f"  概率分布: {probs}")
            print()
        
        print("\n=== Task3 第二部分完成 ===")
        print("逻辑回归分类器构建和评估完成！")
```
